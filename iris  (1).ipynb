```python
x=df.iloc[:,:4]
y=df.iloc[:,4]
x
```

```python
y
```

```python
#Import train_test_split to split the data into train and test datasets.

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#import LogisticRegression from the sci-kit learn library.

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#predicting the results using predict method.
y_pred=model.predict(x_test)
y_pred
```

```python
#import accuracy_score and confusion_matrix from the sci-kit learn metric library.
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```
